```python
import os
import numpy as np

def ds_func(input_directory,final_out):
    for file in os.listdir(input_directory):
        for c in [0,1]:
            # Downsample 'file'
            input_path = os.path.join(input_directory,file)            
            
            image_type = 'ims' # can be ims or Tif

            # Ignored
            outdir = '/nafs/dtward/dong/donglab_resample_lightsheet_good_2022_09_06_tmp'

            !python downsample.py $input_path $image_type $outdir --channel $c

            # Move the downsampled image from 'src_path_xxx' to 'dest_path'
            src_path_jpg = '/ifshome/abenneck/image_registration/donglab_workflows/pipeline_building/'+os.path.splitext(os.path.split(file)[-1])[0]+'_ch_'+str(c)+'_pow_[0.125]_down.jpg'
            src_path_npz = '/ifshome/abenneck/image_registration/donglab_workflows/pipeline_building/'+os.path.splitext(os.path.split(file)[-1])[0]+'_ch_'+str(c)+'_pow_[0.125]_down.npz'

            !mv $src_path_jpg $final_out
            print(f"========== Moved jpg {os.path.splitext(os.path.split(file)[-1])[0]}_ch_{c} ==========")

            !mv $src_path_npz $final_out
            print(f"========== Moved npz {os.path.splitext(os.path.split(file)[-1])[0]}_ch_{c} ==========")

    print()
    print(f'========== Finished processing {input_directory} ==========')
    return
```

```python
in_dir = r'/panfs/dong/3D_registration/Yang_MORF_DragonFly/TME10-1_30x'
final_out = r'/nafs/dtward/dong/final_output/TME10-1'
ds_func(in_dir,final_out)

in_dir = r'/panfs/dong/3D_registration/Yang_MORF_DragonFly/TME10-3_10x'
final_out = r'/nafs/dtward/dong/final_output/TME10-3'
ds_func(in_dir,final_out)

# in_dir = r'/panfs/dong/3D_registration/Yang_MORF_DragonFly/TME10-3_30x'
# final_out = r'/nafs/dtward/dong/final_output/TME10-3'
# ds_func(in_dir,final_out)
```
